```python
import matplotlib.pyplot as plt
from numpy import linspace
from math import sqrt, log, factorial
```

```python
# y = 0n+1 and label the curve 'O(1)'
x1 = list(linspace(0.01,100,100))
y1 = [(0*n)+1 for n in x1]
plt.plot(x1,y1, c = 'g', label = 'O(1)')
# y = log(n) and label the curve 'O(log n)'
y2 = [log(n) for n in x1]
plt.plot(x1,y2, c = 'r', label = 'O(log n)')

# y = n and label it 'O(n)'
y3 = [n for n in x1]
plt.plot(x1,y3, c = 'b', label = 'O(n)')

# y = n*log(n) and label it 'O(nlog(n))'
y4 = [n*log(n) for n in x1]
plt.plot(x1,y4, c = 'tab:gray', label = 'O(nlog(n))')

# y = n^2 and label it 'O(n^2)'
y5 = [n**2 for n in x1]
plt.plot(x1,y5, c = 'm', label = 'O(n^2)')

# y = 2^n and label it 'O(2^n)'
y6 = [2**n for n in x1]
plt.plot(x1,y6, c = 'tab:olive', label = 'O(2^n)')

# y = n! and label it 'O(n!)'
y7 = [factorial(int(n)) for n in x1]
plt.plot(x1,y7, c = 'midnightblue')

# y = n^n and label it 'O(n^n)'
y8 = [n**n for n in x1]
plt.plot(x1,y8, c = 'limegreen', label = 'O(n^n)')

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.ylim(-10,500)
plt.xlim(0,100)
plt.legend(loc='upper right')
```
