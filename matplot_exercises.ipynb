```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1. plot y = x^2 - x + 2

x = np.linspace(-40,40, 10001)
y = x**2-x+2
plt.annotate('Origin', xy = (0,0), xytext = (0,80), arrowprops = {'facecolor': 'green'})
plt.plot(x,y)
```

```python
2. #a. plot y = sqrt(x)
x = np.linspace(0,50,1001)
y = x**.5
plt.plot(x,y)
```

```python
#b. plot y = x**3
x = np.linspace(-50,50,1001)
y = x**3
plt.plot(x,y)
```

```python
#c. plot y = 2**x
x = np.linspace(-50, 50, 1001)
y = 2**x
plt.plot(x,y)
```

```python
#d. plot y = 1/(x+1)
# np.seterr(divide='ignore', invalid='ignore') is what i found to get rid of dividing by zero errors
np.seterr(divide='ignore', invalid='ignore')

x = np.linspace(-5,5,1001)
y = 1/(x+1)
plt.plot(x,y)
```

```python
#3. combine all 4 into subplots
# sqrt(x)
x1 = np.linspace(0,50,1001)
y1 = x**.5
plt.title('y = sqrt(x)')
plt.subplot(221)
plt.plot(x1,y1)
# y = x**3
x2 = np.linspace(-50,50,1001)
y2 = x**3
plt.title('y = x^3')
plt.subplot(222)
plt.plot(x2,y2)
# y = 2**x
x3 = np.linspace(-50, 50, 1001)
y3 = 2**x
plt.title('y = 2^x')
plt.subplot(223)
plt.plot(x3,y3)
#y = 1/(x+1)
x4 = np.linspace(-5,5,1001)
y4 = (1/(x+1))
plt.title('y = 1/(x+1)')
plt.subplot(224)
plt.plot(x4,y4)

plt.show()
```

```python
#4. combine all 4 into subplots with extras
# sqrt(x)
x1 = np.linspace(0,50,1001)
y1 = x**.5
plt.title('y = sqrt(x)')
plt.plot(x1,y1, color = 'green', label = 'sqrt(x)')

# y = x**3
x2 = np.linspace(-50,50,1001)
y2 = x**3
plt.title('y = x^3')
plt.plot(x2,y2, color = 'purple', label = 'x^3') 

# y = 2**x
x3 = np.linspace(-50, 50, 1001)
y3 = 2**x
plt.title('y = 2^x')
plt.plot(x3,y3, color = 'blue', label = '2^x')

#y = 1/(x+1)
x4 = np.linspace(-5,5,1001)
y4 = (1/(x+1))
plt.title('y = 1/(x+1)')
plt.plot(x4,y4, color = 'red', label = '1/(x+1')

plt.legend()

plt.show()
```
