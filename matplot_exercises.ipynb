```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1. plot y = x^2 - x + 2

x = np.linspace(-40,40, 10001)
y = x**2-x+2
plt.annotate('Origin', xy = (0,0), xytext = (0,80), arrowprops = {'facecolor': 'green'})
plt.plot(x,y)
```

```python
2. #a. plot y = sqrt(x)
x = np.linspace(0,50,1001)
y = x**.5
plt.plot(x,y)
```
