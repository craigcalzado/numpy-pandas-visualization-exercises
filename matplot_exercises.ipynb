```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1. plot y = x^2 - x + 2

x = np.linspace(-40,40, 10001)
y = x**2-x+2
plt.annotate('Origin', xy = (0,0), xytext = (0,650), arrowprops = {'facecolor': 'green'})
plt.plot(x,y)
```

```python
2. #a. plot y = sqrt(x)
x = np.linspace(0,50,1001)
y = x**.5
plt.plot(x,y)
```

```python
#b. plot y = x**3
x = np.linspace(-50,50,1001)
y = x**3
plt.plot(x,y)
```

```python
#c. plot y = 2**x
x = np.linspace(-50, 50, 1001)
y = 2**x
plt.plot(x,y)
```

```python
#d. plot y = 1/(x+1)
# np.seterr(divide='ignore', invalid='ignore') is what i found to get rid of dividing by zero errors
np.seterr(divide='ignore', invalid='ignore')

x = np.linspace(-50,50,1001)
y = 1/(x+1)
plt.plot(x,y)
```

```python
#3. combine all 4 into subplots


# sqrt(x)
x = np.linspace(0,50,1001)
y1 = x**.5
plt.subplot(221)
plt.plot(x,y1)
plt.title('y = sqrt(x)')
plt.grid(True, ls = '--')

# y = x**3
x2 = range(0,1001)
y2 = [n**3 for n in x]
plt.subplot(222)
plt.plot(x2,y2)
plt.title('y = x^3')
plt.grid(True, ls = '--')

# y = 2**x
y3 = 2**x
plt.subplot(223)
plt.plot(x,y3)
plt.title('y = 2^x')
plt.grid(True, ls = '--')

#y = 1/(x+1)
y4 = (1/(x+1))
plt.subplot(224)
plt.plot(x,y4)
plt.title('y = 1/(x+1)')
plt.grid(True, ls = '--')

plt.subplots_adjust(wspace=.5, hspace=.5)
```

```python
#4. combine all 4 into subplots with extras
# sqrt(x)
x = np.linspace(-10,
            10,1001)
y1 = x**.5
plt.plot(x,y1, color = 'green', label = 'sqrt(x)')


# y = x**3
x2 = range(0,1001)
y2 = [n**3 for n in x]
plt.plot(x2,y2, color = 'purple', label = 'x^3') 

# y = 2**x
y3 = 2**x
plt.plot(x,y3, color = 'blue', label = '2^x')

#y = 1/(x+1)
y4 = (1/(x+1))
plt.plot(x,y4, color = 'red', label = '1/(x+1')

plt.ylim(-10,10)
plt.xlim(-5,5)

plt.legend()
plt.grid(True)

plt.title('Fun_Functions')
plt.annotate('Origin', xy = (0,0), xytext = (1,-3), arrowprops = {'facecolor': 'green'})
```
